```python
# Import the desk
from google.colab import files
import pandas as pd # Import the pandas library and assign it the alias 'pd'

uploaded = files.upload()  # Sube el archivo CSV desde tu computadora

# Cargar el archivo en un DataFrame
df = pd.read_csv(next(iter(uploaded)))
df.head()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Load the dataset
pd.set_option('display.max_columns', None)
dataset_4 = pd.read_csv('SAT_by_Year_Gender_1967_2001.csv')

# Step 1: Features and Targets
features = ['Year']
targets = ['M_verbal', 'F_verbal', 'M_math', 'F_math']

X = dataset_4[features]
y = dataset_4[targets]

# Step 2: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 3: Train Models
models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree Regressor": DecisionTreeRegressor(random_state=42),
    "Random Forest Regressor": RandomForestRegressor(random_state=42)
}

# Step 4: Evaluate Models
evaluation_results = {}

for model_name, model in models.items():
    # Train the model
    model.fit(X_train, y_train)
    # Predict
    y_pred = model.predict(X_test)
    # Metrics for each target variable
    mse = mean_squared_error(y_test, y_pred, multioutput='raw_values')
    mae = mean_absolute_error(y_test, y_pred, multioutput='raw_values')
    r2 = r2_score(y_test, y_pred, multioutput='raw_values')
    evaluation_results[model_name] = {"MSE": mse.tolist(), "MAE": mae.tolist(), "R2": r2.tolist()}

evaluation_results
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import pandas as pd

# Cargar el dataset
sat_dataset = pd.read_csv("SAT_by_Year_Gender_1967_2001.csv")

# Paso 1: Definir categorías para A_math
def categorize_score(score):
    if score < 500:
        return 'Low'
    elif 500 <= score <= 550:
        return 'Medium'
    else:
        return 'High'

# Aplicar categorización
sat_dataset['A_math_category'] = sat_dataset['A_math'].apply(categorize_score)

# Codificar categorías en etiquetas numéricas
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(sat_dataset['A_math_category'])

# Seleccionar características
X = sat_dataset[['Year', 'M_verbal', 'F_verbal', 'M_math', 'F_math']]

# Dividir en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Entrenar modelo de regresión logística
logistic_model = LogisticRegression(max_iter=500, random_state=42)
logistic_model.fit(X_train, y_train)

# Predicciones
y_pred = logistic_model.predict(X_test)

# Evaluar modelo
accuracy = logistic_model.score(X_test, y_test)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred, target_names=label_encoder.classes_)

# Resultados
print(f"Accuracy: {accuracy}")
print("\nConfusion Matrix:\n", conf_matrix)
print("\nClassification Report:\n", class_report)
```
