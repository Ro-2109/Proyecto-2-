```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import numpy as np
import pandas as pd  # Import pandas for DataFrame operations
from google.colab import files # Import the desk
import io # Import the io module

# Upload the file
uploaded = files.upload()

# Get the filename and file content
filename = next(iter(uploaded))
file_content = uploaded[filename]

# Read the CSV file from the uploaded content using io.BytesIO
dataset = pd.read_csv(io.BytesIO(file_content)) # Assuming your data is in a CSV file and you want to name the DataFrame 'dataset'

# Step 1: Categorize Performance Levels
# Define categories: Low (0-60), Medium (60-80), High (80-100)
def categorize_performance(score):
    if score < 60:
        return 'Low'
    elif score < 80:
        return 'Medium'
    else:
        return 'High'

# Apply categorization based on average score
dataset['Performance'] = dataset['average_score'].apply(categorize_performance)

# Step 2: Feature and Target Selection
features = ['math_score', 'reading_score', 'writing_score']
target = 'Performance'

X = dataset[features]
y = dataset[target]

# Encode target variable
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Step 3: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.3, random_state=42, stratify=y_encoded)

# Step 4: Train a Random Forest Classifier
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

# Step 5: Predictions and Evaluation
y_pred = rf_model.predict(X_test)

# Metrics
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred, target_names=label_encoder.classes_)

accuracy, conf_matrix, class_report
```

# ***Preparación de los Datos***
# ***Definición de las Categorías de Desempeño:***

Crear una nueva variable categórica basada en las puntuaciones promedio:
Alto desempeño: promedio >= 80.
Desempeño medio: 50 <= promedio < 80.
Bajo desempeño: promedio < 50.
División del Conjunto de Datos:

Dividir el dataset en conjuntos de entrenamiento y prueba (80%-20%).
Escalado de Variables:

Normalizar las variables numéricas para mejorar el rendimiento del modelo.
2. Modelos de Clasificación
Entrenar y evaluar los siguientes modelos:
Regresión Logística.
Árbol de Decisión.
Bosques Aleatorios (Random Forest).
3. Evaluación del Modelo
Métricas a utilizar:
Matriz de confusión.
Accuracy, Precision, Recall, y F1-Score.
Curva ROC y AUC.


```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
import numpy as np

# Preparación de datos
# Crear la variable categórica "Performance_Level" basada en el promedio
def categorize_performance(avg_score):
    if avg_score >= 80:
        return "Alto"
    elif avg_score >= 50:
        return "Medio"
    else:
        return "Bajo"

dataset['Performance_Level'] = dataset['average_score'].apply(categorize_performance)

# Dividir en conjuntos de características (X) y etiqueta (y)
X = dataset[['math_score', 'reading_score', 'writing_score']]
y = dataset['Performance_Level']

# Codificar etiquetas para modelos de clasificación
y_encoded = y.map({'Bajo': 0, 'Medio': 1, 'Alto': 2})

# Dividir los datos en conjuntos de entrenamiento y prueba (80%-20%)
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Escalado de características
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Entrenar el modelo de Regresión Logística
log_reg = LogisticRegression(multi_class='multinomial', solver='lbfgs', random_state=42, max_iter=1000)
log_reg.fit(X_train_scaled, y_train)

# Predicciones y evaluación
y_pred = log_reg.predict(X_test_scaled)
y_pred_proba = log_reg.predict_proba(X_test_scaled)

# Generar métricas
conf_matrix = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred, target_names=['Bajo', 'Medio', 'Alto'])

# Curva ROC y AUC (para clasificación multiclase)
fpr = {}
tpr = {}
roc_auc = {}
for i in range(3):  # Para las tres clases
    fpr[i], tpr[i], _ = roc_curve(y_test == i, y_pred_proba[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Resultados
classification_rep, conf_matrix, roc_auc
```

```python
# Mostrar informe de clasificación y matriz de confusión
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar la matriz de confusión
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=['Bajo', 'Medio', 'Alto'], yticklabels=['Bajo', 'Medio', 'Alto'])
plt.title("Matriz de Confusión - Regresión Logística")
plt.xlabel("Predicción")
plt.ylabel("Realidad")
plt.show()

# Mostrar el informe de clasificación
print("Informe de Clasificación:\n", classification_rep)

# Visualización de la curva ROC para cada clase
plt.figure(figsize=(10, 6))
for i in range(3):  # Tres clases
    plt.plot(fpr[i], tpr[i], label=f'Clase {i} (AUC = {roc_auc[i]:.2f})')
plt.plot([0, 1], [0, 1], 'k--', label='Aleatorio')
plt.title('Curva ROC - Regresión Logística')
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```
