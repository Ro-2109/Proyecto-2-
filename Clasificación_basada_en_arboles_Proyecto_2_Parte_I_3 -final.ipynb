Problemática Elegida: Clusterización

Objetivo: Agrupar a los estudiantes según patrones de sueño y actividades diarias.
Justificación:
Identificar grupos puede ayudar a generar recomendaciones específicas sobre hábitos saludables.
Es desafiante debido a las diferencias individuales significativas y la combinación de variables temporales con otras continuas.

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
import pandas as pd

# Cargar el dataset
dataset_3 = pd.read_csv("student_sleep_patterns.csv")

# Selección de características y objetivo
features = [
    'Sleep_Duration', 'Study_Hours', 'Screen_Time', 'Caffeine_Intake',
    'Physical_Activity', 'Weekday_Sleep_Start', 'Weekend_Sleep_Start',
    'Weekday_Sleep_End', 'Weekend_Sleep_End'
]
target = 'Sleep_Quality'

X = dataset_3[features]
y = dataset_3[target]

# Dividir en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Modelos
models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree Regressor": DecisionTreeRegressor(random_state=42),
    "Random Forest Regressor": RandomForestRegressor(random_state=42)
}

# Evaluación
evaluation_results = {}
for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    evaluation_results[model_name] = {"MSE": mse, "MAE": mae, "R2": r2}

# Resultados
print(evaluation_results)
```
